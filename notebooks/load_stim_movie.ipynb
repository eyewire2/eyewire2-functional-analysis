## Load stimuli saved with QDSpy

```python
%load_ext autoreload
%autoreload 2

import pickle
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
STIM_MOV_PATH = Path("../data/stimuli-as-movies")
STIM_MOV_EXT = ".pckle" 
```

```python
def load_qdspy_movie(pickle_path):
    """ Load a pickle file generated by the record-stimulus-as-movie function
        of QDSpy and returns it as a numpy.ndarray
    """
    print(f"Loading pickle file from: {pickle_path}")
    with open(pickle_path, 'rb') as f:
        data = pickle.load(f)
    
    if isinstance(data, np.ndarray):
        image_array = data
        print("Done.")
    else:
        image_array = None
        print("Error: File does not contain a simple numpy array as expected.")    
    
    return image_array
```

```python
tmp_path = Path.joinpath(STIM_MOV_PATH, "RGC_MovingBar" +STIM_MOV_EXT)
mov_DS = load_qdspy_movie(tmp_path)
```

```python
nFr, dx, dy, nCh = mov_DS.shape
mov_flat = np.zeros((dx,dy,nCh), dtype=np.float64) 


mov_range = mov_DS[0:nFr-1]
mov_flat = np.sum(mov_range, axis=0, dtype=np.float64)
mov_flat *= 1.0/mov_flat.max() 

fig, axes = plt.subplots(1, 1, figsize=(12, 3))
   
axes.imshow(mov_flat)
axes.axis('off')
plt.tight_layout()
plt.show()
```
