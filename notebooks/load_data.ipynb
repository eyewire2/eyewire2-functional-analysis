```python
# Check that functional and morphological dataframes can be merged
assert (df_eyewire2_roi_level["date"].astype(str) == id_df["date"].astype(str)).all()
assert (df_eyewire2_roi_level["region"] == id_df["region"]).all()
assert (df_eyewire2_roi_level["field"] == id_df["field"]).all()
assert (df_eyewire2_roi_level["roi_id"].astype(str) == id_df["roi_id"].astype(str)).all()

for (i_new, row_new), (i_old, row_old) in zip(morph_df.iterrows(), id_df.iterrows()):
    xn, yn, zn = tuple(map(float, row_new['Nuc Coords'].split(',')))
    xo, yo, zo = tuple(map(float, row_old['nucleus loc pcg'].split(',')))

    assert np.allclose([xo, yo, zo], [xn, yn, zn])
```
