```python
import os
from matplotlib import pyplot as plt
import pandas as pd
from IPython.core.display_functions import clear_output
```

# Load data

```python
from eyewire2_functional_analysis import data_loader

data_folder = "../../../data/preprocessed-data"
morph_folder = "../../../data/morphological-data"
```

```python
print(sorted(os.listdir(morph_folder)))
```

```python
df_rois_morph = data_loader.load_df_rois_morph(
    morph_folder=morph_folder,
    morph_spreadsheet_filename='Eyewire II Proofread Cells Master List - All Cells 2026-03-17.csv',
    nuc_col_master='Final NucID',
    seg_col_master='Final SegID',
    data_folder=data_folder,
)
```

# Plot data

```python
from eyewire2_functional_analysis.plot_dataframe import plot_df_chirp_and_bar
```

```python
fig_dir = './figures/responses_by_field'
os.makedirs(fig_dir, exist_ok=True)

for (field, cell_class), df_plot in df_rois_morph.groupby(['field', 'Cell Class']):
    clear_output()
    fig, axs = plot_df_chirp_and_bar(
        df=df_plot, id_col='Final NucID', type_col='Cell Type',
        title=f"{field}: All {cell_class}s", chirp_q_thresh=0.45, bar_q_thresh=0.6)
    plt.savefig(os.path.join(fig_dir, f'response_overview_{field}_{cell_class}.pdf'))
    plt.close(fig)
```

```python
fig_dir = './figures/responses_by_type'
os.makedirs(fig_dir, exist_ok=True)

for cell_type, df_plot in df_rois_morph.groupby('Cell Type'):
    clear_output()
    fig, axs = plot_df_chirp_and_bar(
        df=df_plot.sort_values(['field', 'roi_id'], inplace=False),
        id_col='Final NucID', type_col='Cell Type',
        title=cell_type, chirp_q_thresh=0.45, bar_q_thresh=0.6)
    plt.savefig(os.path.join(fig_dir, f'response_overview_{cell_type}.pdf'))
    plt.close(fig)
```
